```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import datetime as dt
import time
```

```python
url = "https://public.coindcx.com/market_data/trade_history?pair=B-BTC_USDT&interval=1m&limit=500"
```

```python
response = requests.get(url)
data = response.json()
df = pd.DataFrame(data)
print(df)
```

```python
df['date_time'] = df['T'].apply(lambda x : time.strftime("%Y-%m-%d %H:%M:%M",time.localtime(int(str(x)[:-3]))))
```

```python
df.reset_index(drop=True)
```

```python
df = df[['date_time','p','q','s','m']]
```

```python

df_final = df.sort_values('p',ascending = False).head(20)
```

```python
df_final.rename(columns = {'date_time':'Date Time','p':'Price','q':'Quantity','s':'Market Name','m':'Market Maker'}, inplace = True)
```
