```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import datetime as dt
import time
```

```python
url = "https://api.wazirx.com/sapi/v1/tickers/24hr"
```

```python
response = requests.get(url)
data = response.json()
df = pd.DataFrame(data)
print(df)
```

```python
df.reset_index(drop=True)
```

```python
df['date_time'] = df['at'].apply(lambda x : time.strftime("%Y-%m-%d %H:%M:%M",time.localtime(int(str(x)[:-3]))))
```

```python
df
```

```python
df = df[['date_time','symbol','baseAsset','quoteAsset','lastPrice']]
```

```python

df_final = df.sort_values('lastPrice',ascending = False).head(20)
```

```python
df_final.rename(columns = {'date_time':'Date Time','symbol':'Symbol','baseAsset':'Base Asset','quoteAsset':'Quote Asset','lastPrice':'Price'}, inplace = True)
```
